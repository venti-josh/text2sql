```python
DDL_PATH = "dbo.ddl"

def get_ddl(table):
    with open(DDL_PATH, "r") as f:
        ddl = [line.replace("\n", "") for line in f.readlines()]
        start = 0
        end = 0
        for i, line in enumerate(ddl):
            stripped_line = line.strip()
            if stripped_line.endswith(f"bizfinway_SIT_STG_TechnoPaints.dbo.{table} ("):
                start = i
            if stripped_line.endswith(");") and start != 0:
                end = i + 1
                break
        return "\n".join(ddl[start:end])
```

```python
def add_table_documentation(model: VannaBase, table: tuple[str, str, str, str], include_description=False, include_usage=False, include_ddl=False):
    if include_description:
        model.add_documentation(f"The table {table[0]} is used for {table[1]}")
    if include_usage:
        model.add_documentation(f"To use {table[0]}, you should {table[2]}")
    if include_ddl:
        model.add_ddl(table[3])
```
